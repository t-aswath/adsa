# **SORTING**



---

Sorting is a fundamental operation in computer science and involves arranging a collection of items in a specific order, such as ascending or descending. There are various sorting algorithms available, each with its own characteristics, advantages, and disadvantages. Here are a few common sorting algorithms in Python:


**1. Bubble Sort:**

**Approach:**
Bubble Sort is a simple comparison-based sorting algorithm that repeatedly steps through the list to be sorted, compares adjacent elements, and swaps them if they are in the wrong order. The pass through the list is repeated until the list is sorted.

- The algorithm compares each pair of adjacent elements and swaps them if they are in the wrong order.
- This process continues until no more swaps are needed, indicating that the list is sorted.

**Time Complexity:**
- Best-case time complexity: O(n) - when the list is already sorted.
- Average-case time complexity: O(n^2) - because it makes n*(n-1)/2 comparisons and swaps in the worst case.
- Worst-case time complexity: O(n^2) - when the list is sorted in reverse order.
- Bubble Sort is not recommended for large datasets due to its poor time complexity.

**2. Selection Sort:**

**Approach:**
Selection Sort divides the input list into two parts: the sorted part and the unsorted part. It repeatedly selects the minimum (or maximum) element from the unsorted part and moves it to the sorted part.

- The algorithm maintains two sublists: the left part is the sorted sublist, and the right part is the unsorted sublist.
- It repeatedly finds the minimum element in the unsorted sublist and moves it to the end of the sorted sublist.

**Time Complexity:**
- Best-case time complexity: O(n^2) - because it makes the same number of comparisons and swaps in all cases.
- Average-case time complexity: O(n^2).
- Worst-case time complexity: O(n^2).

**3. Merge Sort:**

**Approach:**
Merge Sort is a divide-and-conquer algorithm that divides the input list into smaller sublists until each sublist contains only one element. It then merges these sublists to produce a new sorted list.

- The algorithm repeatedly divides the list into two halves until it reaches single-element sublists (base case).
- It then merges the sorted sublists to create larger sorted sublists, until the entire list is sorted.

**Time Complexity:**
- Best-case time complexity: O(n log n) - when the list is divided into approximately equal sublists.
- Average-case time complexity: O(n log n).
- Worst-case time complexity: O(n log n) - Merge Sort maintains a consistent time complexity, making it efficient for large datasets.
- Merge Sort has a stable time complexity for various input cases.

**4. Quick Sort:**

**Approach:**
Quick Sort is another divide-and-conquer algorithm that selects a "pivot" element and partitions the list into two sublists: elements less than the pivot and elements greater than the pivot. It then recursively sorts the sublists.

- The algorithm selects a pivot element from the list (typically the middle element).
- It partitions the list into two sublists: elements less than the pivot and elements greater than the pivot.
- Recursively, Quick Sort is applied to the two sublists.
- The sorted sublists are merged, along with the pivot element.

**Time Complexity:**
- Best-case time complexity: O(n log n) - when the pivot selection leads to well-balanced partitions.
- Average-case time complexity: O(n log n).
- Worst-case time complexity: O(n^2) - when the pivot selection consistently leads to unbalanced partitions.
- Quick Sort has excellent average-case performance and is widely used.

The choice of sorting algorithm depends on factors like the size of the dataset, the desired performance, and whether stability or adaptivity is required. Merge Sort and Quick Sort are often preferred for their better average-case performance, while Bubble Sort and Selection Sort are simpler but less efficient for larger datasets.


## **bubble sort**

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
```

## **selection sort**

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

## **merge sort**

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    left = merge_sort(left)
    right = merge_sort(right)
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

## **quick sort**

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
```

# **Comparison**

```python
import matplotlib.pyplot as plt
import timeit
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    left = merge_sort(left)
    right = merge_sort(right)
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def generate_random_list(n):
    return [random.randint(1, n) for _ in range(n)]

def measure_runtime(sort_func, arr):
    return timeit.timeit(lambda: sort_func(arr.copy()), number=1)

n_values = [100, 200, 300, 400, 500]
bubble_sort_runtimes = []
selection_sort_runtimes = []
merge_sort_runtimes = []
quick_sort_runtimes = []

for n in n_values:
    arr = generate_random_list(n)
    bubble_sort_runtimes.append(measure_runtime(bubble_sort, arr))
    selection_sort_runtimes.append(measure_runtime(selection_sort, arr))
    merge_sort_runtimes.append(measure_runtime(merge_sort, arr))
    quick_sort_runtimes.append(measure_runtime(quick_sort, arr))

plt.figure(figsize=(10, 5))
plt.plot(n_values, bubble_sort_runtimes, marker='o', label='Bubble Sort')
plt.plot(n_values, selection_sort_runtimes, marker='o', label='Selection Sort')
plt.plot(n_values, merge_sort_runtimes, marker='o', label='Merge Sort')
plt.plot(n_values, quick_sort_runtimes, marker='o', label='Quick Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Runtime (seconds)')
plt.title('Sorting Algorithm Comparison')
plt.legend()
plt.grid()

plt.show()
```
